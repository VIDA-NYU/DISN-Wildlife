```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="disn-wild")
```

```python
from transformers import pipeline
classifier = pipeline('zero-shot-classification',
                      model='facebook/bart-large-mnli')
hypothesis_template = 'This product advertisement is about {}.'
```

```python
labels = ["a real animal",
          "a toy",
          "a print of an animal",
          "an object",
          "a faux animal",
          "an animal body part",
          "an faux animal body part"]
```

```python
sequence= "notebook tiger for sale"
classifier(sequence, labels,
           hypothesis_template=hypothesis_template)
```

```python
def get_inf(x: pd.Series) -> List[Dict]:
        label_name = None
        label_description = None
        score_description = None
        score_name = None
        if x["name"] and type(x["name"]) == str:
            res_name = classifier(x["name"], labels)
            label_name = res_name["labels"][0]
            score_name = res_name["scores"][0]
        if x["description"] and type(x["description"]) == str:
            res_description = classifier(x["description"], labels)
            label_description = res_description["labels"][0]
            score_description = res_description["scores"][0]

        x["label_description"] = label_description
        x["score_description"] = score_description
        x["label_name"] = label_name
        x["score_name"] = score_name    
        return x
    
def get_label(x):
    if x["label_name"] == None and x["label_description"] == None:
        return None
    elif x["label_name"] == None :
        return x["label_description"]
    elif x["label_description"] == None :
        return x["label_name"]
    else:
        if x["score_description"] > x["score_name"]:
            return x["label_description"]
        else:
            return x["label_name"]
```
